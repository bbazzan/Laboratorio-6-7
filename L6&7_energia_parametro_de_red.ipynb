```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.optimize import curve_fit
```

```python
E0 = np.array([-27.669221, -30.314085, -31.341254, -31.317018, -30.640438])
pdr = np.array([3.786996, 3.951648, 4.1163, 4.280952, 4.445604])
```

```python
plt.figure(figsize=(21,10))
plt.scatter(pdr, E0)
plt.xlabel('Parametro de red', fontsize=20)
plt.ylabel('E0', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

```python
def parabola(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
popt, pcov = curve_fit(parabola, pdr, E0)
print(popt)
```
