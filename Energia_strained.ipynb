```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#z = np.asarray([3.5, 3.6, 3.7, 3.8, 3.9, 4, 4.1, 4.2, 4.3])
#E = np.asarray([-28.943071, -29.796804, -30.576758, -30.840247, -31.113284, -31.264335, -31.318708, -31.299044, -31.220259])

z = np.asarray([3.7, 3.8, 3.9, 4, 4.1, 4.2, 4.3])
E = np.asarray([-30.576758, -30.840247, -31.113284, -31.264335, -31.318708, -31.299044, -31.220259])

plt.figure(figsize=(10,8))
plt.scatter(z, E)
plt.grid(True)
plt.show()
```

```python
def cuadratica(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
x = np.linspace(3.25, 4.75, 1000)
```

```python
popt, pcov = curve_fit(cuadratica, z, E)
print(popt)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(z, E)
plt.plot(x, cuadratica(x, popt[0], popt[1], popt[2]), color='orange')
plt.grid(True)
plt.show()
```

```python
print(np.where(cuadratica(x, popt[0], popt[1], popt[2]) == min(cuadratica(x, popt[0], popt[1], popt[2]))))
```
